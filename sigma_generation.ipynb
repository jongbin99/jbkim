```python
import os
import pymol
from pymol import cmd
import subprocess
```

```python
mol2_dir = "/Users/JB/Rotation_bkslab/ampC_sigma/docked_poses_sigma"

# Get all .mol2 files in the directory
mol2_files = [f for f in os.listdir(mol2_dir) if f.endswith(".mol2")]

# Convert each .mol2 file to .pdb using Open Babel
for mol2_file in mol2_files:
    mol2_path = os.path.join(mol2_dir, mol2_file)
    pdb_path = os.path.join(mol2_dir, mol2_file.replace(".mol2", ".pdb"))

    try:
        # Run Open Babel conversion command
        subprocess.run(["obabel", mol2_path, "-O", pdb_path], check=True)
        print(f"✅ Converted: {mol2_file} → {pdb_path}")
    except subprocess.CalledProcessError as e:
        print(f"⚠ Error converting {mol2_file}: {e}")

print("\n✅ All MOL2 files have been converted to PDB format.")
```

```python
# Define directories
input_dir = "/Users/JB/Rotation_bkslab/ampC_sigma/predicted_poses_sigma"
output_dir = "/Users/JB/Rotation_bkslab/ampC_sigma/predicted_poses_sigma_pdb"
os.makedirs(output_dir, exist_ok=True)  # Ensure output directory exists

# Get all .cif files
cif_files = [f for f in os.listdir(input_dir) if f.endswith(".cif")]

# Initialize PyMOL in quiet mode
pymol.finish_launching(['pymol', '-qc'])  # '-qc' = Quiet mode (no GUI)

for cif_file in cif_files:
    try:
        cif_path = os.path.join(input_dir, cif_file)
        pdb_name = os.path.splitext(cif_file)[0] + ".pdb"  # Replace .cif with .pdb
        pdb_path = os.path.join(output_dir, pdb_name)

        # Load CIF file into PyMOL
        cmd.reinitialize()  # Reset PyMOL for each structure
        cmd.load(cif_path, "structure")

        # Save as PDB without losing information
        cmd.save(pdb_path, "structure")

    except Exception as e:
        print(f"⚠ Error processing {cif_file}: {e}")

# Quit PyMOL after processing
cmd.quit()

print("\n✅ All CIF files have been converted to PDB format successfully.")
```

```python
ligand_dir = "/Users/JB/Rotation_bkslab/ampC_sigma/docked_poses_sigma"
protein_pdb = "/Users/JB/Rotation_bkslab/ampC_sigma/docked_poses_sigma/sigma2_rec.crg.pdb"
output_dir = "/Users/JB/Rotation_bkslab/ampC_sigma/docked_sigma_complexes"

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Get all ligand PDB files in the directory, excluding the protein itself
ligand_files = [f for f in os.listdir(ligand_dir) if f.endswith(".pdb") and f != "sigma2_rec.crg.pdb"]

# Process each ligand separately using an individual PyMOL script
for ligand_file in ligand_files:
    try:
        ligand_path = os.path.join(ligand_dir, ligand_file)
        output_pdb = os.path.join(output_dir, ligand_file)  # Keep same name as ligand file

        # Corrected PyMOL script (properly formatted)
        pymol_script = f"""\
reinitialize
load {protein_pdb}, protein
load {ligand_path}, ligand
create merged_complex, protein or ligand
save {output_pdb}, merged_complex
quit
"""

        # Write the PyMOL script to a temporary file
        pymol_script_path = os.path.join(output_dir, f"merge_{ligand_file}.pml")
        with open(pymol_script_path, "w") as f:
            f.write(pymol_script)

        # Run PyMOL using the script file
        os.system(f"pymol -qc {pymol_script_path}")

        print(f"✅ Merged PDB saved: {output_pdb}")

        # Remove the temporary PyMOL script after execution
        os.remove(pymol_script_path)

    except Exception as e:
        print(f"⚠ Error processing {ligand_file}: {e}")

print("\n✅ All merged PDB files have been saved in:", output_dir)
```

```python

complexes_dir = "/Users/JB/Rotation_bkslab/ampC_sigma/docked_sigma_complexes"
predicted_dir = "/Users/JB/Rotation_bkslab/ampC_sigma/predicted_poses_sigma_pdb"
output_dir = "/Users/JB/Rotation_bkslab/ampC_sigma/aligned_structures_pred_sigma"
os.makedirs(output_dir, exist_ok=True)  # Ensure output directory exists

# Get all PDB files in both directories
complex_files = {f for f in os.listdir(complexes_dir) if f.endswith(".pdb")}
predicted_files = {f for f in os.listdir(predicted_dir) if f.endswith(".pdb")}

# Find matching filenames in both folders
matching_files = complex_files.intersection(predicted_files)

for pdb_file in matching_files:
    try:
        complex_path = os.path.join(complexes_dir, pdb_file)
        predicted_path = os.path.join(predicted_dir, pdb_file)
        aligned_pdb_path = os.path.join(output_dir, f"aligned_{pdb_file}")

        # Create a PyMOL script for each structure
        pymol_script = f"""
        load {complex_path}, complex
        load {predicted_path}, predicted
        align predicted, complex
        save {aligned_pdb_path}, predicted
        quit
        """

        # Save script as a temporary file
        script_path = os.path.join(output_dir, f"{pdb_file}.pml")
        with open(script_path, "w") as f:
            f.write(pymol_script)

        # Run PyMOL in a separate process
        subprocess.run(["pymol", "-cq", script_path], check=True)

    except subprocess.CalledProcessError as e:
        print(f"⚠ Error aligning {pdb_file}: {e}")

print("\n✅ All matching PDBs have been aligned and saved.")
```
