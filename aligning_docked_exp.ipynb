```python
import os
import subprocess
```

```python
complexes_dir = "/Users/JB/Rotation_bkslab/Mac1_pdb_chain_A"
predicted_dir = "/Users/JB/Rotation_bkslab/Mac1_docked_poses/docked_complexes"
aligned_dir = "/Users/JB/Rotation_bkslab/Mac1_docked_poses/aligned_structures_dock"
ligand_output_dir = "/Users/JB/Rotation_bkslab/Mac1_docked_poses/aligned_ligands_dock"

# Ensure output directories exist
os.makedirs(aligned_dir, exist_ok=True)
os.makedirs(ligand_output_dir, exist_ok=True)

# Get all PDB files in both directories
complex_files = {f for f in os.listdir(complexes_dir) if f.endswith(".pdb")}
predicted_files = {f for f in os.listdir(predicted_dir) if f.endswith(".pdb")}

# Find matching filenames in both folders
matching_files = complex_files.intersection(predicted_files)

for pdb_file in matching_files:
    try:
        complex_path = os.path.join(complexes_dir, pdb_file)
        predicted_path = os.path.join(predicted_dir, pdb_file)
        aligned_pdb_path = os.path.join(aligned_dir, f"aligned_{pdb_file}")
        ligand_output_path = os.path.join(ligand_output_dir, f"ligand_{pdb_file}")

        # Create a PyMOL script for each structure
        pymol_script = f"""
        load {complex_path}, complex
        load {predicted_path}, predicted
        align predicted, complex
        save {aligned_pdb_path}, predicted
        quit
        """

        # Save script as a temporary file
        script_path = os.path.join(aligned_dir, f"{pdb_file}.pml")
        with open(script_path, "w") as f:
            f.write(pymol_script)

        # Run PyMOL in a separate process
        subprocess.run(["pymol", "-cq", script_path], check=True)

        # Extract only LIG_B (ligand) entries from aligned PDB file
        with open(aligned_pdb_path, "r") as f:
            ligand_entries = [line for line in f if " LIG " in line]  # Extract only LIG_B lines

        # Save extracted ligand data
        with open(ligand_output_path, "w") as f:
            f.writelines(ligand_entries)

        print(f"✅ Aligned and extracted LIG_B: {ligand_output_path}")

    except subprocess.CalledProcessError as e:
        print(f"⚠ Error aligning {pdb_file}: {e}")

print("\n✅ All matching PDBs have been aligned and LIG_B coordinates extracted.")
```

```python
# Define directories
complexes_dir = "/Users/JB/Rotation_bkslab/Mac1_pdb_chain_A"
ligand_output_dir = "/Users/JB/Rotation_bkslab/Mac1_pdb_chain_A/ligand_ref"

# Ensure output directory exists
os.makedirs(ligand_output_dir, exist_ok=True)

# Function to extract LIG1 entries from a given PDB file
def extract_ligand_entries(pdb_file, source_dir, output_dir):
    pdb_path = os.path.join(source_dir, pdb_file)
    ligand_output_path = os.path.join(output_dir, f"ligand_{pdb_file}")

    try:
        # Read and filter only lines containing "LIG1"
        with open(pdb_path, "r") as f:
            ligand_entries = [line for line in f if " LIG" in line]

        # Save extracted ligand data if LIG1 is found
        if ligand_entries:
            with open(ligand_output_path, "w") as f:
                f.writelines(ligand_entries)
            print(f"✅ Extracted LIG1 from: {pdb_file}")

    except Exception as e:
        print(f"⚠ Error processing {pdb_file}: {e}")

# Process PDB files in docked_ampc_complexes
pdb_files = [f for f in os.listdir(complexes_dir) if f.endswith(".pdb")]
for pdb_file in pdb_files:
    extract_ligand_entries(pdb_file, complexes_dir, ligand_output_dir)

print("\n✅ All PDB files have been processed. LIG1 entries saved in:", ligand_output_dir)
```

```python
import os
import subprocess
# Define input directory
input_dir = "/Users/JB/Rotation_bkslab/250203_alphafold3/PDBS_lig"

#"/Users/JB/Rotation_bkslab/Mac1_docked_poses/aligned_ligands_dock"
#"/Users/JB/Rotation_bkslab/Mac1_pdb_chain_A/ligand_ref"

# Get all PDB files in the directory
pdb_files = [f for f in os.listdir(input_dir) if f.endswith(".pdb")]

# Loop through each PDB file and deprotonate using Open Babel
for pdb_file in pdb_files:
    file_path = os.path.join(input_dir, pdb_file)

    try:
        # Run Open Babel to remove hydrogens (-d flag removes explicit hydrogens)
        subprocess.run(["obabel", file_path, "-O", file_path, "-d"], check=True)
        print(f"✅ Deprotonated: {pdb_file}")
    except subprocess.CalledProcessError as e:
        print(f"⚠ Error processing {pdb_file}: {e}")

print("\n✅ All PDB files have been deprotonated and updated in:", input_dir)
```

```python
input_dir = "/Users/JB/Rotation_bkslab/ampC_sigma/docked_ligands_sigma"

# Get all PDB files in the directory
pdb_files = [f for f in os.listdir(input_dir) if f.endswith(".pdb")]

# Loop through each PDB file and fix incorrect atomic labels
for pdb_file in pdb_files:
    file_path = os.path.join(input_dir, pdb_file)

    try:
        # Read the PDB file
        with open(file_path, "r") as f:
            lines = f.readlines()

        # Modify only the last column (element names)
        updated_lines = []
        for line in lines:
            if line.startswith(("ATOM", "HETATM")):  # Process atom lines only
                # Ensure line is at least 78 characters long to access last column
                if len(line) >= 78:
                    last_col = line[76:].replace("1+", "").strip()  # Remove '1-'
                    new_line = line[:76] + f"{last_col:>2}\n"  # Maintain formatting
                    updated_lines.append(new_line)
                else:
                    updated_lines.append(line)
            else:
                updated_lines.append(line)  # Keep other lines unchanged

        # Overwrite the original file with corrected data
        with open(file_path, "w") as f:
            f.writelines(updated_lines)

        print(f"✅ Fixed atom labels in: {pdb_file}")

    except Exception as e:
        print(f"⚠ Error processing {pdb_file}: {e}")

print("\n✅ All PDB files have been updated in:", input_dir)
```
