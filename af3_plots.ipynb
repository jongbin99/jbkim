```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import os
import glob
from scipy.stats import pearsonr, linregress
from scipy.stats import gaussian_kde
```

```python
# Load Excel file
file_path = "/Users/JB/Rotation_bkslab/250203_alphafold3/20241209_mac1.xlsx"
data = pd.read_excel(file_path)

# Extract x (Ligand pLDDT) and y (Ligand RMSD) columns
y1 = data.iloc[:, 12]  # Assuming 12th column is Ligand RMSD
x = data.iloc[:, 13]  # Assuming 13th column is Ligand pLDDT
z = data.iloc[:, 18]

# Ensure data is numeric and drop NaN values
df = pd.DataFrame({"Ligand pLDDT": x, "Ligand RMSD": y1}).dropna()
df = df.astype(float)  # Ensure numeric conversion

# Filter data for y2 (where Phe156 backbone RMSD > 156)
#filtered_data = data.loc[(data["Ala129 Backbone RMSD"] > 3) & (data["Gly130 Backbone RMSD"] > 3)]
filtered_data = data.loc[(data["Phe156 Backbone RMSD"] > 1)]
x_filtered = filtered_data.iloc[:, 13]  # Corresponding iPTM values
y2 = filtered_data.iloc[:, 12]  # Corresponding Ligand RMSD values

# Calculate correlation coefficient
correlation_coefficient = np.corrcoef(x, y1)[0, 1]

# Compute confusion matrix components
false_positives = ((x_filtered < 80) & (y2 < 2)).sum()
false_negatives = ((x_filtered > 80) & (y2 > 2)).sum()
true_negatives = ((df["Ligand RMSD"] > 2) & (df["Ligand pLDDT"] < 80)).sum()
true_positives = ((df["Ligand RMSD"] < 2) & (df["Ligand pLDDT"] > 80)).sum()

# Create scatter plot with density map
plt.figure(figsize=(8, 6))

# Plot density map
sns.kdeplot(x=df["Ligand pLDDT"], y=df["Ligand RMSD"], cmap="Blues", fill=True, alpha=0.5, levels=30, thresh=0.1)

# Scatter plot of actual data points
plt.scatter(x, y1, color='blue', alpha=0.5, label=f'All Data (Correlation: {correlation_coefficient:.2f})')
plt.scatter(x_filtered, y2, color='red', alpha=0.8, label="Ala129, Gly130 RMSD > 3")
print(x_filtered)
# Add labels and title
plt.title("Correlation and Density Map of Ligand pLDDT vs. Ligand RMSD")
plt.xlabel("Ligand pLDDT")
plt.ylabel("Ligand RMSD")
plt.xlim(50, 100)
plt.legend()

# Show grid
plt.grid(True, linestyle="--", alpha=0.6)

# Show plot
plt.show()

# Print confusion matrix results
print("Confusion Matrix Counts:")
print(f"False Positives (y > 2 & x > 80): {false_positives}")
print(f"False Negatives (y < 2 & x < 80): {false_negatives}")
print(f"True Negatives (y > 2 & x < 80): {true_negatives}")
print(f"True Positives (y < 2 & x > 80): {true_positives}")

# Print correlation result
print(f"\nPearson Correlation Coefficient: {correlation_coefficient:.4f}")
```

```python
# Load data
file_path = "/Users/JB/Rotation_bkslab/ampC_sigma/ligand_plddt.xlsx"
data = pd.read_excel(file_path)

# Extract IC50, L-pLDDT, and Target columns
x = data.iloc[:, 9]  # IC50 values
y = data.iloc[:, 13]  # L-pLDDT values
category = data.iloc[:, 0].astype(str)  # Target category

# Create DataFrame
df = pd.DataFrame({"IC50": x, "L-pLDDT": y, "Target": category}).dropna()
# Clean IC50 column (remove ">" signs and spaces)
df["IC50"] = df["IC50"].astype(str).str.replace(">", "").str.strip()

# Convert IC50 to numeric, allowing scientific notation
df["IC50"] = pd.to_numeric(df["IC50"], errors="coerce")

# Remove invalid or zero IC50 values (since log(0) is undefined)
df = df[df["IC50"] > 0]

# Compute pKi (-log10(IC50))
df["pKi"] = -np.log10(df["IC50"]*(1e-6))

# Convert L-pLDDT to numeric and drop invalid rows
df["L-pLDDT"] = pd.to_numeric(df["L-pLDDT"], errors="coerce")
df = df.dropna()  # Ensure valid data

# Define color mapping for targets
target_colors = {"Mac1": "blue", "Sigma2": "green", "AmpC": "red", "AmpC non-hits": "orange"}

# Initialize dictionaries to store Pearson correlation and average pKi results
correlation_results = {}
avg_pKi_results = {}

plt.figure(figsize=(8, 6))

# Loop through each target group from our mapping that exists in the data
for target, color in target_colors.items():
    # Filter the DataFrame for this target (using case-insensitive match)
    group_df = df[df["Target"].str.lower() == target.lower()]
    
    if len(group_df) < 2:
        r, p = float("nan"), float("nan")
    else:
        r, p = pearsonr(group_df["pKi"], group_df["L-pLDDT"])
    correlation_results[target] = (r, p)
    
    # Calculate pKi range and count for this target group
    if len(group_df) > 0:
        min_pKi = group_df["pKi"].min()
        max_pKi = group_df["pKi"].max()
        count = len(group_df)
    else:
        min_pKi, max_pKi, count = float("nan"), float("nan"), 0
    pKi_range_results[target] = (min_pKi, max_pKi, count)
    
    # Scatter plot for this group with label including correlation and pKi range/count
    plt.scatter(group_df["pKi"], group_df["L-pLDDT"], alpha=0.7,
                color=color,
                label=(f"{target} (n={count}): r={r:.2f}, "
                       f"p={p:.2e}, pKi range=({min_pKi:.2f}-{max_pKi:.2f})"))

# Labels and title for the plot
plt.xlabel("pKi")
plt.ylabel("L-pLDDT")
plt.title("Correlation between pKi and L-pLDDT by Target")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(loc="upper right", fontsize=9, title="Target Group")
plt.show()

# Print Pearson correlation and pKi range results for each target group
print("Pearson Correlation and pKi Range by Target:")
for target in target_colors.keys():
    r, p = correlation_results.get(target, (float("nan"), float("nan")))
    min_pKi, max_pKi, count = pKi_range_results.get(target, (float("nan"), float("nan"), 0))
    print(f"{target}: Pearson r = {r:.2f}, p = {p:.2e}, Count = {count}, pKi Range = ({min_pKi:.2f}-{max_pKi:.2f})")
```
