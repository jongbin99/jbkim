```python
import os
import pymol
from pymol import cmd
import time
import subprocess

# Define input and output directories
input_dir = "/Users/JB/Rotation_bkslab/Mac1_docked_poses"
output_dir = "/Users/JB/Rotation_bkslab/Mac1_docked_poses/docked_complexes"
os.makedirs(output_dir, exist_ok=True)  # Ensure output directory exists

# Define the protein file
protein_file = os.path.join(input_dir, "mac1_rec.crg.pdb")

ligand_files = [f for f in os.listdir(input_dir) if f.endswith(".mol2")]

# Function to process each ligand
def process_ligand(ligand_file):
    ligand_path = os.path.join(input_dir, ligand_file)
    complex_name = os.path.splitext(ligand_file)[0]  # Remove .mol2 extension
    output_pdb = os.path.join(output_dir, f"{complex_name}.pdb")

    # Construct PyMOL script for ligand processing
    pymol_script = f"""
    load {protein_file}, rec_protein
    load {ligand_path}, dock_ligand
    create merged_complex, rec_protein or dock_ligand
    save {output_pdb}, merged_complex
    quit
    """

    # Save PyMOL script to a temp file
    script_path = os.path.join(output_dir, f"{complex_name}.pml")
    with open(script_path, "w") as f:
        f.write(pymol_script)

    # Run PyMOL in a new subprocess
    try:
        subprocess.run(["pymol", "-cq", script_path], check=True)
    except subprocess.CalledProcessError as e:
        print(f" Error processing {ligand_file}: {e}")

# Process each ligand
for ligand_file in ligand_files:
    process_ligand(ligand_file)

print("\n All complexes have been generated.")
```
